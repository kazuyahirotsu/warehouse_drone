```python
from PIL import Image, ImageDraw
import barcode
from barcode.writer import ImageWriter

# Define A4 size in pixels (300 DPI for better quality)
A4_WIDTH = 2480
A4_HEIGHT = 3508
MARGIN = 100
GAP = 50
BARCODE_HEIGHT = 300  # Fixed height for barcodes

# Create an A4 canvas
canvas = Image.new('RGB', (A4_WIDTH, A4_HEIGHT), 'white')
draw = ImageDraw.Draw(canvas)

# Define the barcode type and data
barcode_type = barcode.get_barcode_class('ean13')
barcode_data = "123456789012"

# Define barcode widths in cm, converted to pixels (300 DPI)
barcode_widths_cm = [3, 4, 6, 7, 8, 9]
barcode_widths_px = [int(width_cm * 118.11) for width_cm in barcode_widths_cm]  # 1 cm = 118.11 pixels at 300 DPI

# Function to generate and place barcode on the canvas
def place_barcode(barcode_width_px, position):
    options = {
        "module_width": barcode_width_px / 1000.0,  # module width in mm (adjust for width)
        "module_height": BARCODE_HEIGHT / 10,  # module height in mm (fixed height)
    }
    barcode_generator = barcode_type(barcode_data, writer=ImageWriter())
    barcode_image = barcode_generator.render(options)
    barcode_image = barcode_image.resize((barcode_width_px, BARCODE_HEIGHT))
    canvas.paste(barcode_image, position)

# Place barcodes in two columns
x_positions = [MARGIN, A4_WIDTH // 2 + MARGIN]
y_position = MARGIN

for barcode_width_px in barcode_widths_px:
    for x in x_positions:
        if y_position + BARCODE_HEIGHT + MARGIN > A4_HEIGHT:
            break
        place_barcode(barcode_width_px, (x, y_position))
    y_position += BARCODE_HEIGHT + GAP

# Save the result
canvas.save('A4_barcodes_varying_widths.png')
```

```python
from PIL import Image, ImageDraw
import barcode
from barcode.writer import ImageWriter
import qrcode
import pdf417gen
from pylibdmtx.pylibdmtx import encode

# Define A4 size in pixels (300 DPI for better quality)
A4_WIDTH = 2480
A4_HEIGHT = 3508
MARGIN = 100
GAP = 50
BARCODE_HEIGHT = 300  # Fixed height for barcodes

# Define the barcode types and data
barcode_types = {
    "EAN-13": barcode.get_barcode_class('ean13'),
    "Code128": barcode.get_barcode_class('code128'),
    "ISBN": barcode.get_barcode_class('isbn'),
    "PZN": barcode.get_barcode_class('pzn'),
    "UPC": barcode.get_barcode_class('upc'),
    "QR Code": "QRCode",
    "PDF417": "PDF417",
    "DataMatrix": "DataMatrix"
}

barcode_data = {
    "EAN-13": "123456789012",
    "Code128": "ABC123456",
    "ISBN": "9781234567890",
    "PZN": "1234567",
    "UPC": "123456789012",
    "QR Code": "https://example.com",
    "PDF417": "Explicit is better than implicit",
    "DataMatrix": "DataMatrix Example"
}

# Define barcode widths in cm, converted to pixels (300 DPI)
barcode_widths_cm = [5, 6, 7, 8, 9, 10]
barcode_widths_px = [int(width_cm * 118.11) for width_cm in barcode_widths_cm]  # 1 cm = 118.11 pixels at 300 DPI

# Function to generate and place barcode on the canvas
def place_barcode(barcode_name, barcode_type, barcode_data, barcode_width_px, position):
    if barcode_name in ["QR Code", "PDF417", "DataMatrix"]:
        if barcode_name == "QR Code":
            qr = qrcode.QRCode(box_size=10)
            qr.add_data(barcode_data)
            qr.make(fit=True)
            barcode_image = qr.make_image(fill='black', back_color='white')
            barcode_image = barcode_image.resize((barcode_width_px, barcode_width_px))
        elif barcode_name == "PDF417":
            codes = pdf417gen.encode(barcode_data, columns=9)
            image = pdf417gen.render_image(codes, scale=3, ratio=3)
            barcode_image = image.resize((barcode_width_px, BARCODE_HEIGHT))
        elif barcode_name == "DataMatrix":
            dm = encode(barcode_data.encode('utf-8'))
            barcode_image = Image.frombytes('RGB', (dm.width, dm.height), dm.pixels)
            barcode_image = barcode_image.resize((barcode_width_px, barcode_width_px))
    else:
        options = {
            "module_width": barcode_width_px / 1000.0,  # module width in mm (adjust for width)
            "module_height": BARCODE_HEIGHT / 10,  # module height in mm (fixed height)
        }
        barcode_generator = barcode_type(barcode_data, writer=ImageWriter())
        barcode_image = barcode_generator.render(options)
        barcode_image = barcode_image.resize((barcode_width_px, BARCODE_HEIGHT))

    canvas.paste(barcode_image, position)

# Generate and save a page for each barcode type
for barcode_name, barcode_type in barcode_types.items():
    # Create an A4 canvas for each type
    canvas = Image.new('RGB', (A4_WIDTH, A4_HEIGHT), 'white')
    draw = ImageDraw.Draw(canvas)

    y_position = MARGIN

    for barcode_width_px in barcode_widths_px:
        x_positions = [MARGIN, A4_WIDTH // 2 + MARGIN]
        for x in x_positions:
            if y_position + BARCODE_HEIGHT + MARGIN > A4_HEIGHT:
                y_position = MARGIN
            place_barcode(barcode_name, barcode_type, barcode_data[barcode_name], barcode_width_px, (x, y_position))
        y_position += BARCODE_HEIGHT + GAP

    # Save the result
    canvas.save(f'A4_barcodes_{barcode_name}_varying_widths.png')
```
